```python
import pandas as pd
import numpy as np
```

```python
def get_df(path):
    with open(path, 'r') as file:
        text = file.read()
        lines = text.split('\n')
        each_line = [l.split(' ') for l in lines[:-1]]
        data = np.array(each_line, dtype=float)
        df = pd.DataFrame(data)
        return df
```

```python
df = get_df('data/DroneTruck-size-100-len-11.txt')
df
```
